```python
import requests
from lxml.html import fromstring
```

```python
url = 'https://www.donga.com/news/Economy/List'
res = requests.get(url, headers = {'User-Agent':'Mozilla/5.0'})
parser = fromstring(res.text)
article_list = []

article_list = parser.xpath('//div[@id="content"]')
parsed_articles = article_list[0].xpath('.//div[@class="rightList"]')
links = []
for article in parsed_articles:
    parsed_link = article.xpath('.//a[@href]')
    link = parsed_link[0].get('href')
    links.append(link)
print(links)
```

```python
articles = [];

for link in links:
    new_url = link;
    res = requests.get(new_url) 
    parser = fromstring(res.text) 
    article_form = parser.xpath('//div[@id="container"]')[0] 
    
    subject = article_form.xpath('.//div[@class="article_title"]/h1/text()')[0] 
    
    article_body = article_form.xpath('.//div[@id="content"]')[0].text_content() 
    article_content = article_body.replace('\xa0','\n') 
    article = {'title' : subject,'link' : new_url, 'content' : article_content} 
    articles.append(article) 

for list in articles:
    print(list)
    print("\m")
```
