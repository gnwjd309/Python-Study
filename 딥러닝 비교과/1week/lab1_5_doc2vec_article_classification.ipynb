#### Doc2vec model
- 명사
- 명사+동사
- parameters
```
window: 모델 학습할때 앞뒤로 보는 단어의 수
vector_size: 벡터 차원의 크기
alpha: learning rate
min_count: 학습에 사용할 최소 단어 빈도 수
epochs: 최대 학습 횟수
```

````
model_nouns = Doc2Vec(df_phone['documents_nouns'].tolist(), vector_size = 100, epochs = 10, window = 7, min_count = 3, workers = 20)
````

````
# model_pos = Doc2Vec(df_phone['documents_pos'].tolist(), vector_size = 100, epochs = 5, window = 7, min_count = 3, workers = 20)
````

````
model_nouns.save('phone_nouns.d2v')
````

````
model_nouns
````

````
# model_pos.save('pos.d2v')
````

#### feature 추출(vector)

````
df_phone['vector_nouns']  = df_phone['tokenize_nouns'].apply(lambda _:model_nouns.infer_vector(_).astype(float))
````
